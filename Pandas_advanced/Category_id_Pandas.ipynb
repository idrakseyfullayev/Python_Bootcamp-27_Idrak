```python
import pandas as pd

# Məlumatların daxil edilməsi
data_100 = {'sütun_adı': ['Bakı', 'Gəncə', 'Bakı', 'Şəki', 'Mingəçevir']}
df_100 = pd.DataFrame(data_100)

# Məlumatları rəqəmlərlə əvəz etmək üçün factorize metodundan istifadə edirik
df_100['rəqəmsal_sütun'] = pd.factorize(df_100['sütun_adı'])[0]

# Rəqəmsal dəyərləri müəyyən aralıqlara bölmək üçün cut metodundan istifadə edirik
# Bu halda, 5 şəhəri 3 kategoriyaya bölürük
df_100['kategorik_sütun'] = pd.cut(df_100['rəqəmsal_sütun'], bins=5, labels=[1,2,3,4,5])

print(df_100)
```
