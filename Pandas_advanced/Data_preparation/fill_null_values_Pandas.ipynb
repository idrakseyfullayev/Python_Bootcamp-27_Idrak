```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

# Məlumatların daxil edilməsi
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': ['a', 'b', 'a', np.nan, 'b']
}
df = pd.DataFrame(data)

# Orta ilə doldurmaq
mean_imputer = SimpleImputer(strategy='mean')
df['A'] = mean_imputer.fit_transform(df[['A']])

# Median ilə doldurmaq
median_imputer = SimpleImputer(strategy='median')
df['B'] = median_imputer.fit_transform(df[['B']])

# Modus ilə doldurmaq
mode_imputer = SimpleImputer(strategy='most_frequent')
df['C'] = mode_imputer.fit_transform(df[['C']])

# KNN ilə doldurmaq (Default n_neighbors=5)
knn_imputer = KNNImputer()
df_knn = pd.DataFrame(knn_imputer.fit_transform(df[['A', 'B']]), columns=['A', 'B'])

print(df)
print(df_knn)
df
```
