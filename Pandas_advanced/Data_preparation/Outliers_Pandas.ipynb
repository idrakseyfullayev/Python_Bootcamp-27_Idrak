```python
def outlier_thresholds(dataframe, col_name, q1=0.25, q3=0.75):
    Q1 = dataframe[col_name].quantile(q1)
    Q3 = dataframe[col_name].quantile(q3)
    IQR = Q3 - Q1
    up_limit = Q3 + 1.5*IQR
    low_limit = Q1 - 1.5*IQR
    return low_limit, up_limit
```

```python
def grab_outliers(dataframe, col_name, index=True):
    low, up = outlier_thresholds(dataframe, col_name)
    
    if dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].shape[0] > 10:
        print(dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].head())
    else:
        print(dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)])
    
    if index:
        outlier_index = dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].index
        return outlier_index
```

```python
def remove_outlier(dataframe, col_name):
    low, up = outlier_thresholds(dataframe, col_name)
    
    df_without_outliers = dataframe[~((dataframe[col_name] < low) | (dataframe[col_name] > up))]
    return df_without_outliers
```

```python
def repalce_with_thresholds(dataframe, col_name):
    low, up = outlier_thresholds(dataframe, col_name)
    
    dataframe.loc[dataframe[col_name] < low, col_name] = low
    dataframe.loc[dataframe[col_name] > up, col_name] = up
    return dataframe
```

### Task


Outlier-ların silinməsi qərarı bir neçə amildən asılıdır və ümumiyyətlə verilən problemlə və verilənlərin analizi məqsədi ilə əlaqədardır. Məsələn:

Analiz Növü: Əgər regresyon kimi model qurmaq istəyirsinizsə, outlier-lar modelin düzgünlüyünə mənfi təsir göstərə bilər və silinməsi tövsiyə oluna bilər.
Verilənlərin Təbiəti: Outlier-lar bəzən məlumatların təbiətinin bir hissəsi ola bilər. Məsələn, maliyyə analizində bəzi böyük məbləğlər normal ola bilər.
Biznes Məqsədi: Əgər biznes məqsədi üçün məlumatları analiz edirsinizsə və outlier-lar analiz nəticələrini əhəmiyyətli dərəcədə dəyişdirirsə, onların silinməsi uyğun ola bilər.
Sizin verdiyiniz nümunə məlumat dəsti kiçikdir və sadəcə iki sütunlu məlumatlardan ibarətdir. Gəlin hər üç metodun nəticələrini təhlil edək və qərar verək:

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.neighbors import LocalOutlierFactor

# Məlumat dəstini yaradın (və ya yükləyin)
data = {
    'feature1': [10, 12, 13, 15, 10, 100, 12, 11, 12, 13],
    'feature2': [100, 150, 200, 250, 300, 1000, 150, 200, 250, 300]
}
df = pd.DataFrame(data)

# 1. Z-score metodu ilə outlier-ları müəyyən etmək
z_scores = np.abs(stats.zscore(df))
z_threshold = 3  # z-score həddi, ümumiyyətlə 3 qəbul edilir
df_zscore = df[(z_scores < z_threshold).all(axis=1)]

# 2. IQR metodu ilə outlier-ları müəyyən etmək
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
iqr_threshold = 1.5  # IQR həddi, ümumiyyətlə 1.5 qəbul edilir
df_iqr = df[~((df < (Q1 - iqr_threshold * IQR)) | (df > (Q3 + iqr_threshold * IQR))).any(axis=1)]

# 3. LOF metodu ilə outlier-ları müəyyən etmək
lof = LocalOutlierFactor(n_neighbors=5, contamination=0.1)
lof_labels = lof.fit_predict(df)
df_lof = df[lof_labels == 1]

print("Orijinal məlumat dəsti:")
print(df)
print("\nZ-score metodu ilə outlier-ların çıxarılması:")
print(df_zscore)
print("\nIQR metodu ilə outlier-ların çıxarılması:")
print(df_iqr)
print("\nLOF metodu ilə outlier-ların çıxarılması:")
print(df_lof)
```

Nəticələrin Təhlili
Z-score Metodu:

Outlier-ların çıxarılması nəticəsində bəzi sətirlər çıxarılacaq.
Nəticələri təhlil edin və yalnız z-score həddini keçən dəyərləri çıxarın.
IQR Metodu:

IQR metodunda müəyyən edilmiş həddlər daxilində olmayan dəyərlər çıxarılacaq.
Bu metod z-score metoduna nisbətən daha konservativdir və məlumat dəstinin təbiətinə daha uyğun ola bilər.
LOF Metodu:

LOF metodu ilə təcrid olunmuş (outlier) nöqtələr çıxarılacaq.
LOF metodu daha mürəkkəbdir və çox zaman daha düzgün nəticələr verir, lakin qismən daha çətindir.
Qərar
Hər üç metoddan əldə olunan nəticələri müqayisə etdikdən sonra, məlumatların təbiətinə və analiz məqsədinə görə hansı metodun ən uyğun olduğuna qərar verin. Əgər məlumatların əhəmiyyətli bir hissəsi outlier olaraq müəyyən edilirsə və bu outlier-lar analiz nəticələrini əhəmiyyətli dərəcədə dəyişdirsə, onları çıxarmaq məqsədəuyğun ola bilər.

Əgər bu addımları atdıqdan sonra daha spesifik bir qərar verə bilmirsinizsə, məlumat dəstinizi daha yaxşı başa düşmək və uyğun addımlar atmaq üçün məlumatın təhlili və vizualizasiyası üzərində daha çox çalışmaq faydalı ola bilər.
