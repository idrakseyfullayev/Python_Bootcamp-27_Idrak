### Case - df1-də olub  df2-də olmayan (Merge ilə) 

```python
import pandas as pd

# 1-ci dataframe (10 sətir, 5 sütun)
df1 = pd.DataFrame({
    'A': range(1, 11),
    'B': list('abcdefghij'),
    'C': range(11, 21),
    'D': range(21, 31),
    'E': range(31, 41)
})

# 2-ci dataframe (eyni data amma 3 sütun çıxarılıb + bəzi sətirlər silinib)
df2 = df1.drop(columns=['C', 'D', 'E']).iloc[:7]  # 7 sətir saxladıq

df1
```

```python
df2
```

```python
result = df1.merge(df2[['B']], on=['B'], how='left', indicator=True)
result
```

```python
only_df1 = result[result['_merge'] == 'left_only']
only_df1
```
