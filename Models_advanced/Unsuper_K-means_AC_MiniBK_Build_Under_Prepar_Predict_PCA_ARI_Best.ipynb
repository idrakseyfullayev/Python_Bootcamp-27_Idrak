## Clip Outliers with IQR Clip method

```python
def boxplot_clip(col):
    Q1 = df.loc[:, col].quantile(0.25)
    Q3 = df.loc[:, col].quantile(0.75)
    IQR = Q3 - Q1
    
    upperbound = Q3 + 1.5 * IQR
    lowerbound = Q1 - 1.5 * IQR
    
    df[col] = df[col].clip(lower=lowerbound, upper=upperbound)
```

```python
for i in df.columns:
    boxplot_clip(i)
```

## K-Means Clustering

### Elbow_methode

```python
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100

from sklearn.cluster import KMeans

wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++', random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
list(enumerate(wcss))
```

## Silhouete_methode

```python

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

range_n_clusters = [2, 3, 4, 5, 6,7,8,9]
for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = clusterer.fit_predict(df)
    silhouette_avg = silhouette_score(df, cluster_labels)
    print("For n_clusters =", n_clusters,"The average silhouette_score is :", silhouette_avg)
```

### 4.Komponentlərin Əhəmiyyətini Öyrənmək (Explained Variance)
Hər bir komponentin məlumatlardakı ümumi dispersiyanın (variance) nə qədərini izah etdiyini görmək üçün "explained variance ratio" dəyərlərinə baxa bilərsiniz.

```python
import matplotlib.pyplot as plt
import numpy as np

# Komponentlərin dispersiyadakı payı
explained_variance = pca.explained_variance_ratio_

# Bu dispersiya faizini yığaraq vizuallaşdırmaq
plt.figure(figsize=(8,6))
plt.plot(np.cumsum(explained_variance), marker='o')
plt.xlabel('Komponent sayı')
plt.ylabel('Yığılan İzah Edilən Variance (%)')
plt.title('PCA ilə İzahlanan Variance Faizinin Qrafiki')
plt.grid(True)
plt.show()

# Hər bir komponentin izahlama faizi:
explained_variance
```

### 6. Komponent Yükləmələri (Feature Contributions)
PCA komponentləri müəyyən atributların hansının daha çox təsir etdiyini göstərə bilər. Bu testdə komponent yükləmələri hesablanır və hər bir atributun hansı komponentə daha çox təsir etdiyini görmək mümkündür.

```python
# Hər atributun komponentlər üzrə yükləmələri (loading)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

# Hər atribut üçün loading göstərmək
loading_df = pd.DataFrame(loadings, columns=[f'PC{i}' for i in range(1, len(loadings[0]) + 1)], index=df_final_pca.columns)
loading_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder obyektini yaradın
label_encoders = {}
for column in dataset.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    dataset[column] = le.fit_transform(dataset[column])
    label_encoders[column] = le
```

```python
# Yuxarıdakı datada X-lər çaxırın kimyəvi göstəriciləri, y isə clusteringdən sonrakı market seqmentasiyasını bildirir. 
# Növbəti mərhələ bütün X-lari daxil etməklə Logistic Regression əsasında y-i predict etməkdir
```

```python
X = dataset.iloc[:, 0:4].values
y = dataset.iloc[:, 4].values
```

```python
X.shape, y.shape
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Applying PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

## explained_variance

```python
explained_variance = pca.explained_variance_ratio_
explained_variance
```

## LogisticRegression Model Build

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# Making the Confusion Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

## Visualising the Training set results

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap

X_set, y_set = X_train, y_train
#minimum və maximum dəyərlərin köməyi ilə x1 və x2 (PCA1 və (PCA2) oxlarının aralığı təyin olunur)
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```
