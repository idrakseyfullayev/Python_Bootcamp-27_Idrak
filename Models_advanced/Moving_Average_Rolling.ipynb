## Moving Average Rolling=5

```python
import pandas as pd

dates = pd.date_range(start='2020-01-01', end='2020-01-31', freq='d').to_list()

numbers = [43, 3, 31, 1, 39, 18, 15, 49, 6, 14, 46, 45, 15, 13, 5, 4, 23, 25, 43, 4, 18, 3, 27, 38, 43, 38, 45, 39,
                 5, 4, 23]

df = pd.DataFrame ({'Dates': dates, 'Price': numbers})
df.head(5)
```

```python
df.plot.line(x='Dates', y='Price')
```

```python
df['Moving_Average'] = df["Price"].rolling(window=5).mean()
```

```python
df.head(10)
```

```python
df.plot.line(x="Dates", y="Moving_Average")
```

```python
df.plot.line(x="Dates", y=["Price", "Moving_Average"])
```

```python
df['Rolling_Average_Center'] = df["Price"].rolling(window=5, center=True).mean()
```

```python
df
```

```python
df
```

```python
df['Min_Period'] = df["Price"].rolling(window=5, min_periods=2).mean()
```

```python
df
```

```python
df.plot.line(x='Dates', y=["Price", "Min_Period"])
```
