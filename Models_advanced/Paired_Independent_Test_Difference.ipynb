```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Random seed for reproducibility
np.random.seed(42)

# Məhsul adları
product_names = [f"Product_{i+1}" for i in range(30)]

# Paired test üçün qiymətlər (Müalicədən əvvəlki və sonra)
pre_treatment_prices = np.random.normal(loc=100, scale=10, size=30)  # Müalicədən əvvəlki qiymətlər
post_treatment_prices = pre_treatment_prices + np.random.normal(loc=5, scale=5, size=30)  # Müalicədən sonrakı qiymətlər

# Independent test üçün qiymətlər
group1_prices = np.random.normal(loc=95, scale=10, size=30)  # Qrup 1 qiymətləri
group2_prices = np.random.normal(loc=110, scale=10, size=30)  # Qrup 2 qiymətləri

# DataFrame yaratma
paired_df = pd.DataFrame({
    'Product': product_names,
    'Pre_Treatment_Price': pre_treatment_prices,
    'Post_Treatment_Price': post_treatment_prices
})

independent_df = pd.DataFrame({
    'Product': product_names[:30],
    'Group1_Price': group1_prices,
    'Group2_Price': group2_prices
})

# Paired t-test
paired_t_stat, paired_p_value = stats.ttest_rel(pre_treatment_prices, post_treatment_prices)

# Independent t-test
independent_t_stat, independent_p_value = stats.ttest_ind(group1_prices, group2_prices)

# Nəticələri çap edək
print("Paired t-test üçün DataFrame:")
print(paired_df.head())  # İlk 5 sətri göstər

print("\nIndependent t-test üçün DataFrame:")
print(independent_df.head())  # İlk 5 sətri göstər

# Test nəticələrini çap edək
print("\nPaired t-test nəticələri:")
print("t-statistic:", paired_t_stat)
print("p-value:", paired_p_value)

print("\nIndependent t-test nəticələri:")
print("t-statistic:", independent_t_stat)
print("p-value:", independent_p_value)
```
