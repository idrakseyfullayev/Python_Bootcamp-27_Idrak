```python
def outlier_thresholds(dataframe, col_name, q1=0.25, q3=0.75):
    Q1 = dataframe[col_name].quantile(q1)
    Q3 = dataframe[col_name].quantile(q3)
    IQR = Q3 - Q1
    up_limit = Q3 + 1.5*IQR
    low_limit = Q1 - 1.5*IQR
    return low_limit, up_limit
```

```python
outlier_thresholds(dataframe=dataset, col_name='MaxTemp')
```

```python
def grab_outliers(dataframe, col_name, index=True):
    low, up = outlier_thresholds(dataframe, col_name)
    
    if dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].shape[0] > 10:
        print(dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].head())
    else:
        print(dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)])
    
    if index:
        outlier_index = dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].index
        return outlier_index
```

```python
grab_outliers(dataframe=dataset, col_name='MaxTemp')
```

```python
def remove_outlier(dataframe, col_name):
    low, up = outlier_thresholds(dataframe, col_name)
    
    df_without_outliers = dataframe[~((dataframe[col_name] < low) | (dataframe[col_name] > up))]
    return df_without_outliers    
```

```python
remove_outlier(dataframe=dataset, col_name='MaxTemp')
```

```python
def repalce_with_thresholds(dataframe, col_name):
    low, up = outlier_thresholds(dataframe, col_name)
    
    dataframe.loc[dataframe[col_name] < low, col_name] = low
    dataframe.loc[dataframe[col_name] > up, col_name] = up
    return dataframe
```

```python
repalce_with_thresholds(dataframe=dataset, col_name='MaxTemp')
```

```python
grab_outliers(dataframe=dataset, col_name='MaxTemp')
```

```python
import pandas as pd  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from scipy.stats import pearsonr

data = pd.read_csv(r"C:\Users\dell\Desktop\Python\week4\day1\Weather.csv")
dataset = data.copy()

# Data
x = dataset['MinTemp'].values.reshape(-1,1)
y = dataset['MaxTemp'].values.reshape(-1,1)

# Train-test bölünməsi
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1, shuffle=True)

# Model təlimi
regressor = LinearRegression().fit(x_train, y_train)
y_pred = regressor.predict(x_test)

# Metriklərin hesablamaq
R2 = r2_score(y_test, y_pred)
print('R^2:', R2)

n = x_test.shape[0]
p = x_test.shape[1]
adjusted_R2 = 1 - (1-R2)*(n-1)/(n-p-1)
print('Adjusted R^2:', adjusted_R2)

MAE = round(mean_absolute_error(y_test, y_pred), 2)
print("MAE:", MAE)

MSE = round(mean_squared_error(y_test, y_pred), 2)
print("MSE:", MSE)

RMSE = round(np.sqrt(mean_squared_error(y_test, y_pred)), 2)
print("RMSE:", RMSE)

# Pearson korelyasiya əmsalını hesablamaq
correlation_matrix = np.corrcoef(y_test.flatten(), y_pred.flatten())
pearson_corr = correlation_matrix[0, 1]
print('Pearson Korelyasiya Əmsalı:', pearson_corr)

# Alternativ olaraq, scipy istifadə etməklə
pearson_corr, _ = pearsonr(y_test.flatten(), y_pred.flatten())
print('Pearson Korelyasiya Əmsalı (scipy ilə):', pearson_corr)

# Intercept və Slope tapmaq
intercept = regressor.intercept_[0]
slope = regressor.coef_[0][0]
print("Intercept (kəsik):", intercept)
print("Slope (meyl əmsalı):", slope)

# Xətti tənliyi qurmaq
print(f"Xətti tənlik: y = {slope} * x + {intercept}")

# Yeni feature
x_1 = 23
print(f'Yeni feature x_1 = {x_1}')

# Predict
print(f"Predict: y = {slope * x_1 + intercept}")
```
