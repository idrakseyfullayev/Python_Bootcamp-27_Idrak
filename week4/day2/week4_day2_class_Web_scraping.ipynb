```python
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    print(i.find('a').text if i.find('a').text.startswith('P') else '')
```

```python
import re
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    if re.match('^P', i.find('a').text):
        print(i.find('a').text)
```

```python
import re
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    if re.findall('^P', i.find('a').text) == ['P']:
        print(i.find('a').text)
```

```python
import re
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    if re.search('sql', i.find('a').text, re.IGNORECASE):
        print(i.find('a').text)
```

### Telim adlari

```python
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    print(i.find('a').text)
```

```python
telim_adlari = []
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    telim_adlari.append(i.find('a').text)
```

```python
telim_adlari
```

### Telim linkleri

```python
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    print(i.find('a').get('href'))
```

```python
telim_linkleri = []
for i in soup.find_all(name='div', attrs={'class': 'news-title'}):
    telim_linkleri.append('http://www.qss.az/' + i.find('a').get('href'))
```

```python
telim_muddetleri = []
for i in soup.find_all(name='div', attrs={'class': 'col-lg-5 col-md-5 col-sm-5 col-xs-5 single-star-b padding0'}):
    telim_muddetleri.append(i.text)
```

```python
telim_muddetleri
```

```python
for i in range(len(telim_muddetleri)):
    try:
        telim_muddetleri[i] = int(telim_muddetleri[i].rstrip(' saat'))
    except ValueError:
        telim_muddetleri[i] = 0
```
