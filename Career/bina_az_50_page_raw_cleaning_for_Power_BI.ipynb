```python
import copy
df = copy.deepcopy(data)
df
```

## 1. Fix the columns

```python
df.columns = df.columns.str.lower()
df.columns
```

### 2. Check the Data Type

```python
df.info()
```

### location column problems

```python
df[['location']].value_counts()
```

```python
df['location'].unique()
```

```python
filtered_data = df[df['location'].str.contains('Nəriman|Gənclik|Qara|Neftçilər|20 Yanvar|Azadlıq', 
                                                                                           case=False, na=False)]
```

```python
filtered_data[['location']].value_counts()
```

### area column problems

```python
filtered_data
```

```python
pd.options.display.float_format = '{:,.1f}'.format
filtered_data.loc[:, 'area'] = filtered_data.area.map(lambda x: x.replace(' m\u00b2', ''))
```

```python
filtered_data.area = filtered_data.area.astype('float')
```

```python
filtered_data
```

```python
filtered_data.info()
```

### floor column problems

```python
filtered_data
```

```python
import re
filtered_data.loc[:, 'floor'] = filtered_data['floor'].apply(lambda x: re.match(r'(\d+)', x).group(1))
```

```python
filtered_data.floor = filtered_data['floor'].astype('int')
```

```python
filtered_data
```

```python
filtered_data.info()
```

### price column problems

```python
filtered_data
```

```python
filtered_data.loc[:, "price"]  = filtered_data.price.apply(lambda x: x.replace(' ', ''))
```

```python
filtered_data.price = filtered_data.price.astype(int)
```

```python
filtered_data
```

```python
filtered_data.info()
```
