```python
# Lazımi kitabxanaları import edirik
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Grafik ölçülərini təyin edirik (eni: 6, hündürlüyü: 3 düym)
plt.figure(figsize=(6, 3))

# Korrelyasiya matrisinin yuxarı üçbucağı üçün maska yaradırıq:
# `np.ones_like` tam 1-lərdən ibarət matris yaradır.
# `np.triu` funksiyası isə matrisin yuxarı üçbucağını saxlayır, qalan hissələri sıfır edir.
mask = np.triu(np.ones_like(filtered_data_num.corr(numeric_only=True), dtype='bool'))

# İki dəyişən arasındakı əlaqəni göstərən istilik xəritəsi yaradırıq:
# `filtered_data.corr(numeric_only=True)` - yalnız rəqəm sütunlarının korrelyasiyasını hesablayır.
# `round(..., 1)` - korrelyasiya dəyərlərini 1 onluq rəqəmədək yuvarlaqlaşdırır.
# `mask=mask` - yuxarı üçbucağı gizlədir (maskanı tətbiq edir).
# `vmin=-1, vmax=1` - rəng xəritəsində korrelyasiya dəyərlərinin minimum və maksimum intervalını müəyyən edir.
# `annot=True` - hər hücrədə dəyərləri göstərir.
# `cmap='BrBG'` - rəng palitrası olaraq "Brown to Green" seçir.
heatmap = sns.heatmap(round(filtered_data_num.corr(numeric_only=True), 1), 
                      mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')

# İstilik xəritəsinə başlıq əlavə edirik:
# `fontsize=18` - başlıq üçün şrift ölçüsünü təyin edir.
# `pad=16` - başlıq ilə istilik xəritəsi arasındakı məsafəni tənzimləyir.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize': 18}, pad=16)

# Qrafiki göstərir:
plt.show()
```

### Outlier problems

```python
filtered_data
```

```python
import pandas as pd

# Hər kateqoriya üçün IQR diapazonunda clip etmək
def clip_outliers(group):
    Q1 = group['price'].quantile(0.25)
    Q3 = group['price'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # `clip` metodunu tətbiq et
    group['price'] = group['price'].clip(lower=lower_bound, upper=upper_bound)
    return group

# Kateqoriyalara görə qrupla və yalnız 'value' sütunu ilə işləyərək nəticəni birləşdir
clipped_df = filtered_data.groupby('location', group_keys=False).apply(lambda group: clip_outliers(group))
# clipped_df = a.groupby('location', group_keys=False).apply(clip_outliers)

clipped_df
```
