```python
loans = {
    'Cus_id': [1, 2, 3, 4],
    'Credit_amount': [232, 342, 656, 123],
    'Branch_id': [32, 23, 43, 54]
}

loans = pd.DataFrame(data=loan)
loans
```

```python
branchs = {
    'Branch_id': [32, 23, 54, 43],
    'Branch_name': ['Baku', "Sumg", "Gence", "Naxc"]
}

branchs = pd.DataFrame(data=branchs)
branchs
```

```python
port = {
    'Cus_id': [1, 4],
    'Rest_money': [13, 43]
}
ports = pd.DataFrame(data=port)
ports
```

```python
pd.merge(left=loans, right=branchs, how='left', on='Branch_id')
```

```python
pd.merge(left=loans, right=branchs, how='left', left_on='Branch_id', right_on='Branch_id')
```

```python
loans.join(branchs, how='left', lsuffix='_l', rsuffix='_b')
```

```python
loans.join(branchs.set_index('Branch_id'), how='left', on='Branch_id')
```

```python
loans.set_index('Branch_id').join(branchs.set_index('Branch_id'))
```

```python
loans.join(branchs.set_index('Branch_id'), how='left', on='Branch_id', validate='m:1')
```
