### IQR methodu

```python
df1 = df.select_dtypes(include=[np.number]).drop(columns=['Credit_History', 'Loan_Amount_Term'])
df1.fillna(df1.mean(), inplace=True)
df1
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR

lower_fence, upper_fence
```

```python
(df1>lower_fence) & (df1<upper_fence) # asagidaki script-i anlamaq ucun
```

```python
df1[((df1>lower_fence) & (df1<upper_fence)).all(axis=1)]
```

```python
df1[((df1<lower_fence) | (df1>upper_fence)).any(axis=1)]
```

```python
df.drop(index=df1[((df1<lower_fence) | (df1>upper_fence)).any(axis=1)].index)
```
