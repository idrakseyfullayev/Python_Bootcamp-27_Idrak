```python
numeric_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
categorical_features_Label = ['education', 'workclass', 'relationship'] # ordinal columns 
categorical_features_OneHot = ['sex', 'marital_status', 'race'] # nominal columns
categorical_features_Fre = ['occupation'] # nominal but category manyit is nominal, 
                                                                #but because it has many categorical values
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, FunctionTransformer

# OneHotEncoder üçün pipeline (Nominal dəyişənlər üçün)
categorical_pipeline_OneHot = Pipeline(
    steps=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'))  # Naməlum dəyərləri yoxlayır
    ])

# LabelEncoder üçün funksiya (Ordinal dəyişənlər üçün)
def encode_labels(X):
    le = LabelEncoder()
    # Her bir sütun üçün LabelEncoder tətbiqi
    return X.apply(lambda col: le.fit_transform(col) if col.dtype == 'object' else col)

# Pipeline-i yaradın
categorical_pipeline_Label = Pipeline(
    steps=[
        ('label', FunctionTransformer(encode_labels))  # Müstəqil funksiyanın tətbiqi
    ]
)

# Frequency encoding funksiyası
def frequency_encoding(X):
    for col in X.columns:
        freqs = X[col].value_counts(normalize=True)
        X[col] = X[col].map(freqs)
    return X

# Frequency encoding funksiyasını Pipeline-a daxil etmək üçün FunctionTransformer istifadə edirik
frequency_encoder = FunctionTransformer(frequency_encoding)

# Pipeline yaradılması
categorical_pipeline_Frequency = Pipeline(
    steps=[
        ('frequency_encode', frequency_encoder)
    ])

# numeric datani scale ucun
numeric_pipeline = Pipeline(
    steps=[
    ('scaler', RobustScaler())
    ])
```

## Columns Transfer

```python
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numeric_features),
        ('cat_labal', categorical_pipeline_Label, categorical_features_Label),
        ('cat_OneHot', categorical_pipeline_OneHot, categorical_features_OneHot),
        ('cat_Fre', categorical_pipeline_Frequency, categorical_features_Fre),
    ])

mypipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('model', LogisticRegression(random_state=42))
                            
    ])
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
mypipeline.fit(X_train, y_train)
mypipeline.score(X_test,y_test)
```

```python
y_pred_logistic = mypipeline.predict(X_test)
```

## Metrics

```python
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score, \
                                                    confusion_matrix, classification_report
print('confiion martrix: \n', confusion_matrix(y_test, y_pred_logistic))
print()
print('accuracy_score', accuracy_score(y_test, y_pred_logistic))
print('roc auc score: ', roc_auc_score(y_test, y_pred_logistic))
print('recall score: ', recall_score(y_test, y_pred_logistic))
print('precesion score: ', precision_score(y_test, y_pred_logistic))
print('f1 score: ', f1_score(y_test, y_pred_logistic))
print()
print('classification report:\n\n', classification_report(y_test, y_pred_logistic))
```
