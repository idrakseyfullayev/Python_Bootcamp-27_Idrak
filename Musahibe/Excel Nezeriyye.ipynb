## vlookup excel

Excel-də `VLOOKUP` funksiyası müəyyən bir dəyəri bir siyahada və ya cədvəldə tapmaq və həmin dəyərin başqa bir sütundakı müvafiq məlumatını qaytarmaq üçün istifadə olunur. Məsələn, bir cədvəldə məhsul koduna görə məhsulun adını tapmaq üçün istifadə edilə bilər.

### `VLOOKUP` sintaksisi:
```excel
=VLOOKUP(lookup_value, table_array, col_index_num, [range_lookup])
```

#### Parametrlər:
1. **`lookup_value`** – Axtarmaq istədiyin dəyər (məsələn, məhsul kodu).
2. **`table_array`** – Axtardığın dəyəri və qaytarılacaq məlumatı ehtiva edən cədvəl (məsələn, məhsul kodu və məhsul adlarının olduğu cədvəl).
3. **`col_index_num`** – Cədvəldə axtarılan dəyəri tapdıqdan sonra hansı sütundakı məlumatı qaytaracağını göstərən sütun nömrəsi. (Birinci sütun 1-dir.)
4. **`range_lookup`** (opsional) – Yaxın təxmini (TRUE) və ya dəqiq uyğunluğu (FALSE) təyin edir.
   - **TRUE** (və ya boş buraxılsa): Yaxın təxmin tapır (dəyərlər artan sırayla sıralanmalıdır).
   - **FALSE**: Dəqiq uyğunluq tapır.

### Məsələn:
Deyək ki, sənin A2:B6 aralığında belə bir cədvəlin var:

| Məhsul Kodu | Məhsul Adı |
|-------------|------------|
| 1001        | Şirniyyat   |
| 1002        | Çörək       |
| 1003        | Süd         |
| 1004        | Meyvə       |

İndi sən, məsələn, **məhsul koduna** görə müvafiq məhsul adını tapmaq istəyirsən. Əgər məhsul kodu **1003**-dürsə, nəticə "Süd" olmalıdır.

```excel
=VLOOKUP(1003, A2:B6, 2, FALSE)
```

Bu formul:
- **1003**-ü axtarır,
- **A2:B6** aralığında,
- Nəticəni **2-ci sütundan** (məhsul adı sütunu) qaytarır,
- **FALSE** istifadə edilir, çünki dəqiq uyğunluq istəyirik.

### Nəticə:
`Süd` olacaq.

Bu şəkildə `VLOOKUP` funksiyası ilə bir siyahıdan və ya cədvəldən məlumatları sürətlə tapa bilərsən.

## INDEX və MATCH

Excel-də `INDEX` və `MATCH` funksiyaları birləşdirilərək, `VLOOKUP` funksiyasından daha çevik bir axtarış yaratmaq mümkündür. `INDEX MATCH` kombinasiyası daha sürətli və dinamikdir, çünki sol tərəfdən sağa deyil, hər iki istiqamətdə axtarış edə bilir və çox sayda məlumatla işləyərkən daha effektivdir.

### `INDEX` funksiyasının sintaksisi:
```excel
=INDEX(array, row_num, [column_num])
```

- **`array`** – Axtarılacaq məlumatların olduğu aralıq.
- **`row_num`** – Məlumatı tapmaq istədiyin sətir nömrəsi.
- **`column_num`** (opsional) – Həmin aralıq daxilində sütun nömrəsi (opsionaldır, bir çox hallarda lazım olmur).

### `MATCH` funksiyasının sintaksisi:
```excel
=MATCH(lookup_value, lookup_array, [match_type])
```

- **`lookup_value`** – Axtardığın dəyər.
- **`lookup_array`** – Dəyəri axtarmaq istədiyin aralıq.
- **`match_type`** – Axtarma növü (adətən 0 olur, bu, dəqiq uyğunluğu göstərir).

### `INDEX` və `MATCH` birlikdə:
`MATCH` funksiyası axtarış üçün sətir və ya sütun nömrəsini qaytarır, `INDEX` funksiyası isə bu sətir/sütun nömrəsinə uyğun məlumatı çıxarır.

### Məsələn:
Aşağıdakı cədvəldə, məhsul koduna əsaslanaraq məhsul adını tapmaq istəyirsən:

| Məhsul Kodu | Məhsul Adı |
|-------------|------------|
| 1001        | Şirniyyat   |
| 1002        | Çörək       |
| 1003        | Süd         |
| 1004        | Meyvə       |

Məsələn, **məhsul kodu 1003**-ün uyğun **məhsul adını** tapmaq üçün `INDEX MATCH` funksiyasını istifadə edə bilərsən:

1. **`MATCH`** funksiyası ilə məhsul kodunun hansı sətirdə olduğunu tap:
   ```excel
   =MATCH(1003, A2:A5, 0)
   ```
   Bu, 1003-ün **3-cü sətirdə** olduğunu qaytaracaq.

2. **`INDEX`** funksiyası ilə həmin sətirdəki məhsul adını qaytar:
   ```excel
   =INDEX(B2:B5, MATCH(1003, A2:A5, 0))
   ```

Bu formul:
- **B2:B5** aralığında məhsul adını qaytarır.
- **MATCH(1003, A2:A5, 0)** vasitəsilə məhsul kodunun yerləşdiyi sətiri tapır (bu sətir **3** olacaq).
- Nəticədə, **"Süd"** dəyərini qaytarır.

### Nəticə:
`Süd` olacaq.

### `INDEX MATCH` üstünlükləri:
1. **İki istiqamətli axtarış:** `VLOOKUP` yalnız sol tərəfdən sağa doğru axtarır, lakin `INDEX MATCH` həm sağdan sola, həm də soldan sağa axtara bilir.
2. **Effektivlik:** Böyük məlumat dəstləri ilə işləyərkən daha sürətli olur.
3. **Dinamik sütunlar:** Əgər sütunlar yerini dəyişərsə, `INDEX MATCH` nəticələrini dəyişməyəcək, ancaq `VLOOKUP` problemi yarada bilər.

Bu şəkildə `INDEX MATCH` kombinasiyası Excel-də çox güclü və çevik axtarış imkanı yaradır.

## sumifs ve sumif ferqi

`SUMIF` və `SUMIFS` funksiyaları Excel-də şərtə əsasən cəmləmə əməliyyatını yerinə yetirirlər, lakin onların əsas fərqləri aşağıdakı kimidir:

1. **SUMIF**:
   - Tək bir şərtə əsasən cəmləmə əməliyyatını yerinə yetirir.
   - Sintaksisi: 
     ```excel
     =SUMIF(range, criteria, [sum_range])
     ```
   - **range**: Şərtə uyğun qiymətlərin axtarılacağı aralıq.
   - **criteria**: Şərt, yəni hansı meyara görə cəmləmə ediləcək.
   - **sum_range**: Cəmlənəcək faktiki rəqəmlərin yerləşdiyi aralıq (istəyə bağlıdır, əgər göstərilməzsə, ilk aralıq üzrə cəmlənəcək).

   **Misal**:
   ```excel
   =SUMIF(A1:A10, ">5", B1:B10)
   ```
   A1-dən A10-a qədər olan hüceyrələrdə 5-dən böyük dəyərlər varsa, həmin dəyərlərə uyğun B1-dən B10-a qədər olan hüceyrələr cəmlənir.

2. **SUMIFS**:
   - Çoxsaylı şərtlərə əsasən cəmləmə əməliyyatını yerinə yetirir.
   - Sintaksisi: 
     ```excel
     =SUMIFS(sum_range, criteria_range1, criteria1, [criteria_range2, criteria2], ...)
     ```
   - **sum_range**: Cəmlənəcək rəqəmlərin yerləşdiyi aralıq.
   - **criteria_range1**: Birinci şərtə uyğun qiymətlərin axtarılacağı aralıq.
   - **criteria1**: Birinci şərt.
   - İkinci və daha çox şərtlər əlavə oluna bilər.

   **Misal**:
   ```excel
   =SUMIFS(B1:B10, A1:A10, ">5", C1:C10, "<10")
   ```
   A1-dən A10-a qədər olan hüceyrələrdə 5-dən böyük, C1-dən C10-a qədər olan hüceyrələrdə isə 10-dan kiçik olanlara uyğun B1-dən B10-a qədər hüceyrələr cəmlənir.

**Nəticə**: 
- `SUMIF` tək şərt ilə işləyir, `SUMIFS` isə çox şərt ilə.

## VLOOKUP ve HLOOKUP feri

Excel-də **VLOOKUP** və **HLOOKUP** funksiyaları məlumat axtarışı üçün istifadə olunur, lakin bu funksiyalar məlumatın necə təşkil olunmasına görə fərqlənir.

### 1. **VLOOKUP (Vertical Lookup)**:
- **Məlumatları şaquli olaraq (sütunlar üzrə) axtarır**.
- Əsasən bir cədvəldə sol tərəfdə yerləşən bir dəyəri tapıb, həmin dəyər ilə eyni sətrdə olan başqa bir sütundakı dəyəri qaytarmaq üçün istifadə edilir.
  
**Sintaksis:**
```excel
=VLOOKUP(lookup_value, table_array, col_index_num, [range_lookup])
```
- **lookup_value**: Axtarmaq istədiyiniz dəyər.
- **table_array**: Məlumatın yerləşdiyi aralıq.
- **col_index_num**: Axtarış dəyəri tapıldıqda hansı sütundakı dəyəri qaytarmaq istədiyiniz sütun nömrəsi.
- **range_lookup**: Dəqiq uyğunluq (`FALSE`) və ya yaxın uyğunluq (`TRUE`).

**Nümunə:**
```excel
=VLOOKUP("İdrak", A2:D10, 3, FALSE)
```
Bu misalda, "İdrak" adlı dəyər A sütununda axtarılır və həmin sətrdəki 3-cü sütunun dəyəri qaytarılır.

---

### 2. **HLOOKUP (Horizontal Lookup)**:
- **Məlumatları üfüqi olaraq (sətirlər üzrə) axtarır**.
- Əsasən bir cədvəldə yuxarıda yerləşən bir dəyəri tapıb, həmin dəyər ilə eyni sütunda olan başqa bir sətirdəki dəyəri qaytarmaq üçün istifadə edilir.

**Sintaksis:**
```excel
=HLOOKUP(lookup_value, table_array, row_index_num, [range_lookup])
```
- **lookup_value**: Axtarmaq istədiyiniz dəyər.
- **table_array**: Məlumatın yerləşdiyi aralıq.
- **row_index_num**: Axtarış dəyəri tapıldıqda hansı sətirdəki dəyəri qaytarmaq istədiyiniz sətir nömrəsi.
- **range_lookup**: Dəqiq uyğunluq (`FALSE`) və ya yaxın uyğunluq (`TRUE`).

**Nümunə:**
```excel
=HLOOKUP("İdrak", A1:D5, 3, FALSE)
```
Bu misalda, "İdrak" adlı dəyər A1-dən D1-ə qədər üfüqi olaraq axtarılır və həmin sütundakı 3-cü sətrin dəyəri qaytarılır.

---

### Əsas Fərqlər:
| Xüsusiyyət           | **VLOOKUP**                   | **HLOOKUP**                 |
|----------------------|-------------------------------|-----------------------------|
| **Axtarışın istiqaməti** | Sütunlar üzrə (şaquli)        | Sətirlər üzrə (üfüqi)        |
| **İstifadə sahəsi**   | Axtarış dəyəri sol sütunda olur, nəticə isə başqa bir sütunda qaytarılır | Axtarış dəyəri yuxarıdakı sətrdə olur, nəticə isə başqa bir sətirdə qaytarılır |

### Hansını istifadə etməli?
- Məlumatınız **şaquli** olaraq sütunlarda təşkil olunubsa, **VLOOKUP** istifadə edin.
- Məlumatınız **üfüqi** olaraq sətirlərdə təşkil olunubsa, **HLOOKUP** istifadə edin.

## XLOOKUP

Bəli, **XLOOKUP** funksiyası Excel-də mövcuddur və **Office 365** və **Excel 2021** versiyalarında təqdim edilmişdir. Bu funksiya daha əvvəlki **VLOOKUP** və **HLOOKUP** funksiyalarını əvəz etməklə yanaşı, onların məhdudiyyətlərini də aradan qaldırır.

---

### **XLOOKUP Funksiyasının Sintaksisi**
```excel
=XLOOKUP(lookup_value, lookup_array, return_array, [if_not_found], [match_mode], [search_mode])
```

### **Parametrlər:**
1. **`lookup_value`** - Axtardığınız dəyər.
2. **`lookup_array`** - Axtarış aparacağınız sütun və ya sıra.
3. **`return_array`** - Uyğun gələn dəyəri qaytarmaq üçün istifadə olunan sütun və ya sıra.
4. **`[if_not_found]`** *(opsional)* - Axtarış uğursuz olarsa, qaytarılacaq mesaj.
5. **`[match_mode]`** *(opsional)* - Axtarış rejimi:
   - 0: Tam uyğunluq (default).
   - -1: Tam uyğunluq; tapılmadıqda ən kiçik növbəti dəyər.
   - 1: Tam uyğunluq; tapılmadıqda ən böyük əvvəlki dəyər.
   - 2: Joker işarələrlə uyğunluq (*, ?).
6. **`[search_mode]`** *(opsional)* - Axtarış istiqaməti:
   - 1: İlk dəyərdən başlayaraq (default).
   - -1: Son dəyərdən başlayaraq.
   - 2: İkili axtarış (artan sıralama).
   - -2: İkili axtarış (azalan sıralama).

---

### **XLOOKUP Funksiyasının Üstünlükləri:**
1. **Dinamik Aralıqlar:** XLOOKUP həm soldan sağa, həm də sağdan sola axtarış edə bilir. VLOOKUP isə yalnız soldan sağa axtarış edə bilir.
2. **Daha Sürətli Axtarış:** Axtarış nəticəsi tapılmazsa, `#N/A` əvəzinə istifadəçi dostu bir mesaj göstərmək mümkündür.
3. **Cüt Sətirlər və ya Sütunlar:** XLOOKUP həm sətir, həm də sütun aralıqlarında işləyə bilir (HLOOKUP və VLOOKUP birlikdə bunu edə bilmir).
4. **Joker Simvolları Dəstəkləyir:** Axtarışda `*` və `?` kimi simvolları istifadə etmək mümkündür.

---

### **Misal:**
#### Problem: Məhsul kodu üzrə qiyməti tapmaq
| Məhsul Kodu | Məhsul Adı   | Qiymət |
|-------------|--------------|--------|
| A001        | Telefon      | 500    |
| A002        | Planşet      | 300    |
| A003        | Noutbuk      | 1000   |

**Formula:**
```excel
=XLOOKUP("A002", A2:A4, C2:C4, "Tapılmadı")
```
**Nəticə:**
> 300

---

Əgər XLOOKUP funksiyası sizin Excel versiyanızda yoxdursa, bu funksiya əvəzinə **INDEX** və **MATCH** kombinasiya funksiyalarından istifadə edə bilərsiniz.
