

Aşağıda hər bir tələb olunan kateqoriyadan 10 vacib funksiya və metodun siyahısı verilmişdir:

---

### **Python: String funksiyaları**
1. `.strip()` - Boşluqları və ya simvolu başdan və sondan silir.
2. `.lower()` - Bütün hərfləri kiçik edir.
3. `.upper()` - Bütün hərfləri böyük edir.
4. `.find(sub)` - Alt simvolun ilk mövqeyini tapır.
5. `.replace(old, new)` - Alt simvolu dəyişdirir.
6. `.split(sep)` - Simvolu verilən ayırıcı ilə bölür.
7. `.join(iterable)` - Iterabledakı elementləri birləşdirir.
8. `.startswith(prefix)` - Simvolun müəyyən prefiks ilə başlayıb-başlamadığını yoxlayır.
9. `.endswith(suffix)` - Simvolun müəyyən suffiks ilə bitib-bitmədiyini yoxlayır.
10. `.count(sub)` - Alt simvolun neçə dəfə təkrarlandığını sayır.

---

### **Python: List funksiyaları**
1. `.append(x)` - Siyahıya element əlavə edir.
2. `.extend(iterable)` - Siyahıya digər iterable elementlərini əlavə edir.
3. `.insert(index, x)` - Müəyyən mövqeyə element daxil edir.
4. `.remove(x)` - Siyahıdan ilk `x` dəyərini silir.
5. `.pop(index)` - Verilən indeksi və ya son elementi silib qaytarır.
6. `.clear()` - Siyahını boşaldır.
7. `.index(x)` - Elementin indeksini qaytarır.
8. `.count(x)` - Siyahıda neçə dəfə təkrarlandığını qaytarır.
9. `.sort()` - Siyahını artan sırada sıralayır.
10. `.reverse()` - Siyahını tərsinə çevirir.

---

### **Python: Dict funksiyaları**
1. `.keys()` - Bütün açarları qaytarır.
2. `.values()` - Bütün dəyərləri qaytarır.
3. `.items()` - (açar, dəyər) cütlərini qaytarır.
4. `.get(key, default)` - Açar mövcuddursa dəyər, yoxdursa default qaytarır.
5. `.pop(key)` - Açarı silir və dəyərini qaytarır.
6. `.update(other)` - Digər lüğətin cütlərini əlavə edir.
7. `.setdefault(key, default)` - Açar yoxdursa, default dəyər əlavə edir.
8. `.copy()` - Lüğətin surətini yaradır.
9. `.clear()` - Bütün açar-dəyərləri silir.
10. `len(dict)` - Lüğətin cütlərinin sayını qaytarır.

7. **`intersection(other_set)`** - İki yığımın kəsişməsini qaytarır.

### **Python: Built-in funksiyalar**
1. **`print(value, ...)`** - Verilənləri ekrana çap edir.  
2. **`len(obj)`** - Verilən obyektin (məsələn, string, list) elementlərinin sayını qaytarır.  
3. **`type(obj)`** - Obyektin məlumat tipini qaytarır.  
4. **`input(prompt)`** - İstifadəçidən məlumat daxil etməyi tələb edir.  
5. **`sum(iterable, start=0)`** - İterable obyektin elementlərinin cəmini qaytarır.  
6. **`min(iterable, *[, key])`** - İterable obyektin ən kiçik elementini qaytarır.  
7. **`max(iterable, *[, key])`** - İterable obyektin ən böyük elementini qaytarır.  
8. **`sorted(iterable, *, key=None, reverse=False)`** - İterable obyektin elementlərini sıralayıb qaytarır.  
9. **`range(start, stop[, step])`** - Müəyyən interval daxilində ardıcıl rəqəmlər yaradır.  
10. **`round(number[, ndigits])`** - Verilən rəqəmi müəyyən dəqiqlikdə yuvarlaqlaşdırır.  
11. **`zip(*iterables)`** - Verilən iterable obyektləri birləşdirərək tuple-lərdən ibarət iterator yaradır.  
12. **`filter(function, iterable)`** - Şərtə uyğun gələn elementləri seçərək yeni iterator yaradır.  
13. **`map(function, iterable)`** - Verilən funksiyanı hər elementə tətbiq edərək yeni iterator yaradır.  
14. **`abs(x)`** - Verilən rəqəmin mütləq qiymətini qaytarır.  
15. **`pow(x, y[, z])`** - Rəqəmin qüvvətini hesablayır (modul parametri varsa, modula görə nəticə qaytarır).  
16. **`all(iterable)`** - İterable obyektin bütün elementləri `True` olduqda `True` qaytarır.  
17. **`any(iterable)`** - İterable obyektin hər hansı elementi `True` olduqda `True` qaytarır.  
18. **`enumerate(iterable, start=0)`** - İterable obyektin elementlərini indekslərlə birlikdə qaytarır.  
19. **`reversed(seq)`** - Verilən sıra obyektini əksinə çevirərək iterator qaytarır.  
20. **`id(obj)`** - Verilən obyektin unikal identifikatorunu qaytarır.  
21. **`hash(obj)`** - Verilən obyektin hash dəyərini qaytarır (mutable obyektlərdə işləməz).  
22. **`bin(x)`** - Verilən rəqəmin ikilik (binary) təqdimatını qaytarır.  
23. **`oct(x)`** - Verilən rəqəmin səkkizlik (octal) təqdimatını qaytarır.  
24. **`hex(x)`** - Verilən rəqəmin onaltılıq (hexadecimal) təqdimatını qaytarır.  
25. **`eval(expression)`** - Verilən string kimi ifadəni icra edir.  
26. **`exec(object)`** - Dinamik olaraq Python kodunu icra edir.  
27. **`open(file, mode)`** - Fayl açmaq üçün istifadə olunur.  
28. **`isinstance(obj, class)`** - Bir obyektin müəyyən bir sinifdən olub-olmadığını yoxlayır.  
29. **`dir([obj])`** - Verilən obyektin atributlarını və metodlarını siyahı şəklində qaytarır.  
30. **`help([obj])`** - Python obyektləri haqqında kömək və ya sənədləşmə göstərir.

### 🔹 **SQL Əsas Əməliyyatlar və Funksiyalar**  

| **Əməliyyat / Funksiya** | **Təsviri** | **Nümunə** |
|-------------------------|------------|------------|
| `SELECT` | Cədvəldən məlumatları seçmək üçün istifadə olunur. | `SELECT * FROM employees;` |
| `INSERT INTO` | Cədvələ yeni məlumat əlavə edir. | `INSERT INTO employees (name, age) VALUES ('Ali', 30);` |
| `UPDATE` | Mövcud məlumatları dəyişdirir. | `UPDATE employees SET age = 31 WHERE name = 'Ali';` |
| `DELETE` | Cədvəldən məlumat silir. | `DELETE FROM employees WHERE age < 25;` |
| `WHERE` | Şərt tətbiq etmək üçün istifadə olunur. | `SELECT * FROM employees WHERE age > 30;` |
| `JOIN` | İki və ya daha çox cədvəli birləşdirir. | `SELECT orders.id, customers.name FROM orders JOIN customers ON orders.customer_id = customers.id;` |
| `GROUP BY` | Məlumatları müəyyən sahəyə görə qruplaşdırır. | `SELECT department, COUNT(*) FROM employees GROUP BY department;` |
| `ORDER BY` | Nəticələri müəyyən sütuna əsasən sıralayır. | `SELECT * FROM employees ORDER BY age DESC;` |
| `LIMIT` | Nəticələrin sayını məhdudlaşdırır. | `SELECT * FROM employees LIMIT 5;` |
| `COUNT()` | Seçilmiş sətirlərin sayını qaytarır. | `SELECT COUNT(*) FROM employees;` |

| `MIN(column)` | Verilmiş sütunda ən kiçik dəyəri qaytarır. | `SELECT MIN(salary) FROM employees;` |
| `MAX(column)` | Verilmiş sütunda ən böyük dəyəri qaytarır. | `SELECT MAX(salary) FROM employees;` |
| `HAVING` | `GROUP BY` ilə birlikdə şərt tətbiq edir. | `SELECT department, COUNT(*) FROM employees GROUP BY department HAVING COUNT(*) > 10;` |
| `DISTINCT` | Unikal dəyərləri seçir. | `SELECT DISTINCT department FROM employees;` |
| `CASE` | Şərti ifadələr yaratmaq üçün istifadə olunur. | `SELECT name, salary, CASE WHEN salary > 5000 THEN 'High' ELSE 'Low' END AS salary_level FROM employees;` |
| `LIKE` | Verilmiş şablona uyğun sətirləri seçir. | `SELECT * FROM employees WHERE name LIKE 'A%';` |
| `IN` | Bir neçə dəyəri yoxlamaq üçün istifadə olunur. | `SELECT * FROM employees WHERE department IN ('IT', 'HR');` |
| `BETWEEN` | Müəyyən aralıqda olan dəyərləri seçir. | `SELECT * FROM employees WHERE age BETWEEN 25 AND 35;` |
| `EXISTS` | Alt sorğunun nəticə qaytarıb-qaytarmadığını yoxlayır. | `SELECT * FROM employees WHERE EXISTS (SELECT 1 FROM departments WHERE department_id = employees.department_id);` |

| `AND` | Bütün şərtlər doğru olarsa, `TRUE` qaytarır. | `SELECT * FROM employees WHERE age > 25 AND salary > 5000;` |
| `OR` | Ən azı bir şərt doğru olarsa, `TRUE` qaytarır. | `SELECT * FROM employees WHERE age > 25 OR salary > 5000;` |
| `NOT` | Şərtin əksini qaytarır. | `SELECT * FROM employees WHERE NOT (age > 25);` |
| `XOR` | Şərtlərdən yalnız biri doğru olduqda `TRUE` qaytarır (MySQL). | `SELECT * FROM employees WHERE age > 25 XOR salary > 5000;` |
| `CASE` | Şərti ifadələr yaratmaq üçün istifadə olunur. | `SELECT name, salary, CASE WHEN salary > 5000 THEN 'High' ELSE 'Low' END AS salary_level FROM employees;` |
| `IF(condition, true_value, false_value)` | Şərt doğru olduqda `true_value`, əks halda `false_value` qaytarır (MySQL). | `SELECT IF(age > 30, 'Older', 'Younger') FROM employees;` |
| `IFNULL(expression, alt_value)` | Əgər ifadə `NULL` olarsa, alternativ dəyəri qaytarır. | `SELECT IFNULL(phone, 'No phone') FROM employees;` |
| `NULLIF(expression1, expression2)` | Əgər iki ifadə eyni olarsa, `NULL`, əks halda `expression1` qaytarır. | `SELECT NULLIF(10, 10); -- NULL` |
| `COALESCE(expression1, expression2, ...)` | İlk `NULL` olmayan dəyəri qaytarır. | `SELECT COALESCE(NULL, NULL, 'Default', 'Another'); -- 'Default'` |
| `NVL(expression, alt_value)` | Əgər ifadə `NULL` olarsa, alternativ dəyəri qaytarır (Oracle). | `SELECT NVL(phone, 'No phone') FROM employees;` |
| `NVL2(expression, value_if_not_null, value_if_null)` | İfadə `NULL` deyilsə, `value_if_not_null`, əks halda `value_if_null` qaytarır (Oracle). | `SELECT NVL2(phone, 'Has Phone', 'No Phone') FROM employees;` |
| `DECODE(expression, search1, result1, search2, result2, ..., default_value)` | `CASE` funksiyasına bənzəyir. Əgər `expression` `searchX` dəyərinə bərabərdirsə, müvafiq `resultX` qaytarır. Əgər uyğunluq tapılmazsa, `default_value` qaytarır (Oracle). | `SELECT DECODE(department, 'IT', 'Information Technology', 'HR', 'Human Resources', 'Other') FROM employees;` |
| `EXISTS(subquery)` | Alt sorğu nəticə qaytararsa, `TRUE`, əks halda `FALSE` qaytarır. | `SELECT * FROM employees WHERE EXISTS (SELECT 1 FROM departments WHERE department_id = employees.department_id);` |

Agregat funksiyalar çoxlu sətirdən ibarət məlumatları bir nəticəyə gətirmək üçün istifadə olunur.

| **Funksiya** | **Təsviri** | **Nümunə** |

### 🔹 **Excel Axtarış və İstinad Funksiyaları (Lookup & Reference Functions)**

| **Funksiya** | **Təsviri** | **Nümunə** |
|--------------|-------------|------------|
| `VLOOKUP()` | Cədvəlin sol tərəfindəki məlumatlardan müəyyən bir məlumatı axtarır və sağ tərəfdəki dəyəri qaytarır. | `=VLOOKUP(102, A2:B10, 2, FALSE)` |
| `HLOOKUP()` | Yatay olaraq cədvəlin bir hissəsində məlumat axtarır. | `=HLOOKUP(102, A1:F10, 3, FALSE)` |
| `INDEX()` | Cədvəldə müəyyən satır və sütun kəsişməsindən dəyəri qaytarır. | `=INDEX(A1:C10, 3, 2)` |

| `INDIRECT()` | Mətnin göstərişi ilə verilən hüceyrə istinadını qaytarır. | `=INDIRECT("A1")` |
| `XLOOKUP()` | Verilən axtarış dəyərini cədvəlin istənilən yerində axtarır və uyğun dəyəri qaytarır (yeni versiyalar). | `=XLOOKUP(102, A2:A10, B2:B10)` |

---

**XLOOKUP()** Excel-in ən yeni funksiyalarından biridir və `VLOOKUP`-dan fərqli olaraq, daha çox çevikdir. `XLOOKUP` həm yatay, həm də şaquli axtarışları asanlaşdırır, habelə axtarış qiyməti tapılmadıqda alternativ bir dəyər qaytarmaq imkanı verir.

### **XLOOKUP Nümunəsi:**
```excel
=XLOOKUP(102, A2:A10, B2:B10, "Not Found")
```
Bu funksiya 102 qiymətini `A2:A10` aralığında axtarır və taparsa, uyğun olan `B2:B10` aralığındakı dəyəri qaytarır. Əgər tapmazsa, "Not Found" mətnini qaytarır.
