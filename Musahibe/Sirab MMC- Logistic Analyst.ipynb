#### ✅ Müsahibəlik cavab

**“Missing value-ları handle edərkən əvvəlcə səbəbini və payını analiz edirəm, sonra uyğun metod seçirəm:”**

1️⃣ **Azdırsa (məs: <5%)**

2️⃣ **Numerical sütunlarda**

5️⃣ **Çoxdursa (məs: >40-50%)**
→ sütunu ümumiyyətlə silməyi düşünürəm

6️⃣ **Advanced hallarda**
→ interpolation, model-based imputation istifadə edirəm

**“Outlier-ləri həm vizual, həm də statistik metodlarla tapıram:”**

1️⃣ **Vizual**

* Boxplot

  → paylanmada kənar dəyərləri görürəm

2️⃣ **IQR metodu (ən çox istifadə edirəm)**

* Q1, Q3 tapıram
* IQR = Q3 - Q1
* Outlier:
  **< Q1 - 1.5×IQR** və ya **> Q3 + 1.5×IQR**

3️⃣ **Z-score**

* |z| > 3 → outlier hesab edirəm

4️⃣ **Business baxışı**

* Məntiqsiz dəyərləri yoxlayıram
  (məs: mənfi satış, absurd delivery time və s.)

#### ✅ 1. “Son 1 ayda satışlar” sorğusu

```sql
SELECT SUM(sales_amount) AS total_sales
FROM sales
WHERE sale_date >= CURRENT_DATE - INTERVAL '1 month';
```

👉 **Qısa izah:** son 30 günün satışlarını filtr edib cəmləyirəm

#### ✅ 2. “Ən çox satış edən məhsul”

```sql
SELECT product_id, SUM(sales_amount) AS total_sales
FROM sales
GROUP BY product_id
ORDER BY total_sales DESC
LIMIT 1;
```

👉 **Qısa izah:** məhsul üzrə qrupla → ən böyüyü seç

---

#### ✅ 3. Moving average necə hesablanır?

```sql
SELECT 
    sale_date,
    AVG(sales_amount) OVER (
        ORDER BY sale_date 
        ROWS BETWEEN 6 PRECEDING AND CURRENT ROW
    ) AS moving_avg_7_days
FROM sales;
```

👉 **Qısa izah:** window function ilə son 7 günün ortalaması

#### ✅ 4. RANK vs DENSE_RANK fərqi

  (1, 2, 2, 4)

* **DENSE_RANK()** → boşluq yoxdur
  (1, 2, 2, 3)

👉 **Killer cümlə:**

#### On-time delivery (OTD) nədir?

Əla, bunu həm **Excel**, həm də **Power BI** üçün necə hesablamağı əlavə edək 👇

---

#### 1️⃣ Excel-də OTD hesablaması

Tutaq ki, sənin datan belədir:

| OrderID | PlannedDate | ActualDate |
| ------- | ----------- | ---------- |
| 101     | 01-04-2026  | 01-04-2026 |
| 102     | 01-04-2026  | 03-04-2026 |
| 103     | 02-04-2026  | 02-04-2026 |

#### Formula:

**Yeni sütun əlavə et: “OnTime”**

```excel
=IF(ActualDate <= PlannedDate, 1, 0)
```

* 1 → vaxtında
* 0 → gecikib

**Ümumi OTD %:**

```excel
=SUM(OnTime)/COUNTA(OrderID)*100
```

✅ Bu, OTD faizini verir

---

#### 2️⃣ Power BI-də OTD hesablaması (DAX)

Tutaq ki, **Orders** cədvəlində `PlannedDate` və `ActualDate` var.

**Step 1 – Yeni ölçü (Measure) yarat:**

```DAX
OTD % = 
DIVIDE(
    COUNTROWS(FILTER(Orders, Orders[ActualDate] <= Orders[PlannedDate])),
    COUNTROWS(Orders),
    0
) * 100
```

* `FILTER` → vaxtında çatdırılan sifarişləri seçir
* `COUNTROWS` → ümumi və vaxtında sifarişləri sayır
* `DIVIDE` → faiz hesablayır

##### 🔹 Route Optimization – Hesablama + Vizuallaşdırma

#### 1️⃣ Hesablama

**Məqsəd:** Hər sifarişi çatdırmaq üçün ən qısa və ya ən sürətli marşrutu tapmaq.

#### **Excel-də**

1. Sifarişlər və depo lokasiyalarının məlumatlarını topla: `OrderID`, `CustomerAddress`, `Depot`, `PlannedTime`, `Distance`.
2. **Distance və Travel Time hesabla:**

   * `Distance` = Excel-də `Google Maps API` və ya `Haversine formula` ilə hesablana bilər:

   ```excel
   =ACOS(SIN(lat1)*SIN(lat2)+COS(lat1)*COS(lat2)*COS(lon2-lon1))*6371
   ```

   * `TravelTime` = `Distance / AverageSpeed`
3. **Optimal marşrut tapmaq:**

   * Excel-də `Solver` ilə **Total Distance** və ya **Total Time** minimallaşdırılır.
   * Məqsəd: bütün sifarişlər üçün minimal yol və vaxt.

#### **Power BI-də (DAX + Power Query)**

1. Sifariş və depo məlumatlarını idxal et.
2. Yeni ölçü / sütun yarat:

```DAX
DistanceKM = // Haversine formula ilə məsafə
TravelTimeHrs = Orders[DistanceKM] / 40   // ortalama sürət 40 km/saat
```

3. Marşrut performansını hesablamaq üçün:

```DAX
TotalDistancePerRoute = SUMX(FILTER(Orders, Orders[RouteID]=EARLIER(Orders[RouteID])), Orders[DistanceKM])
AverageDeliveryTime = AVERAGEX(FILTER(Orders, Orders[RouteID]=EARLIER(Orders[RouteID])), Orders[TravelTimeHrs])
```

**Power BI-də / Excel-də göstərmək üçün:**

#### Warehouse efficiency necə ölçülür?

Bu terminlər logistika və supply chain (təchizat zənciri) göstəriciləridir. Sadə dildə mənaları belədir:

   * Anbardan sifarişlərin toplanması və paketlənməsi prosesində işçi və ya sistemin nə qədər effektiv işləməsi.

### Forecast (proqnoz) analizi

Excel-də **forecast (proqnoz) analizi** əsasən keçmiş dataya baxaraq gələcək dəyərləri hesablamaq üçün istifadə olunur. Sən bunu 2 əsas üsulla edə bilərsən:

   * 1 sütun → **dəyər (satış, tələb və s.)**

2. Data-nı seç

3. Menü:
   👉 **Data → Forecast Sheet**

4. Açılan pəncərədə:

   * Chart növü seç (line / column)
   * Forecast end date (nə vaxta qədər proqnoz)
   * Confidence interval (etibarlılıq intervalı)

#### Nəticə:

#### 🔹 2. **Formula ilə Forecast (daha professional)**

---

#### **1️⃣ On-time Delivery (OTD)**

**Məna:** Sifarişlərin vaxtında çatdırılma faizi.

**Formula:**
$$
\text{OTD (%)} = \frac{\text{Vaxtında çatdırılan sifarişlər}}{\text{Ümumi sifarişlər}} \times 100
$$

#### Excel:

* Yeni sütun yarat: `Vaxtında çatdırıldı?` → 1 = vaxtında, 0 = gecikdi.
* Hesablamaq üçün:

```excel
=SUM([Vaxtında çatdırıldı?])/COUNT([Sifariş ID])*100
```

* Vizuallıq: `Clustered Column` və ya `Card` vizual ilə OTD göstərə bilərsən.

#### Power BI:

* Yeni Measure:

```DAX
OTD (%) = DIVIDE(
    COUNTROWS(FILTER(Sales, Sales[DeliveryOnTime]=1)),
    COUNTROWS(Sales)
) * 100
```

* Vizuallıq: `Card` vizual + `Clustered Column` / `Gauge` ilə göstərə bilərsən.

---

#### **2️⃣ Delivery Time (Orta çatdırılma müddəti)**

**Məna:** Sifarişlərin çatdırılma müddətinin orta dəyəri (günlə).

**Formula:**
$$
\text{Delivery Time (days)} = \text{Delivery Date} - \text{Order Date}
$$

### Excel:

* Yeni sütun əlavə et:

```excel
= [Delivery Date] - [Order Date]
```

* Orta:

```excel
=AVERAGE([Delivery Time])
```

* Vizuallıq: `Line Chart` zamanla orta çatdırılma göstərmək üçün, `Histogram` isə dağılım üçün.

#### Power BI:

* Yeni Measure:

```DAX
Avg Delivery Time = AVERAGE(Sales[DeliveryDate] - Sales[OrderDate])
```

* Vizuallıq: `Line Chart` zamanla və ya `Column Chart`.

---

#### **3️⃣ Cost per Delivery (Bir sifarişin xərci)**

**Formula:**
$$
\text{Cost per Delivery} = \frac{\text{Total Delivery Cost}}{\text{Total Deliveries}}
$$

#### Excel:

```excel
= SUM([Delivery Cost]) / COUNT([Order ID])
```

* Vizuallıq: `Column Chart`, `Line Chart` və ya `Card` vizualı.

#### Power BI:

* Measure:

```DAX
Cost per Delivery = DIVIDE(SUM(Sales[DeliveryCost]), COUNTROWS(Sales))
```

* Vizuallıq: `Card` və `Line Chart` (trend üçün).

---

#### **4️⃣ Inventory Turnover (Anbar dövriyyəsi)**

**Formula:**
$$
\text{Inventory Turnover} = \frac{\text{COGS (Cost of Goods Sold)}}{\text{Average Inventory}}
$$

#### Excel:

```excel
= [COGS] / AVERAGE([Inventory])
```

* Vizuallıq: `Line Chart` zamanla və ya `Column Chart`.

#### Power BI:

* Measure:

```DAX
Inventory Turnover = DIVIDE(SUM(Sales[COGS]), AVERAGE(Inventory[StockLevel]))
```

* Vizuallıq: `Line Chart` və ya `Column Chart`.

---

#### **5️⃣ Fill Rate (Sifarişlərin tamamlanma faizi)**

**Məna:** Sifarişlərin neçə faizi tam qarşılanır.

**Formula:**
$$
\text{Fill Rate (%)} = \frac{\text{Tam qarşılanan sifarişlər}}{\text{Ümumi sifarişlər}} \times 100
$$

### Excel:

* Yeni sütun: `Fullfilled?` → 1 = tam qarşılandı, 0 = qismən və ya yox.

```excel
=SUM([Fullfilled?])/COUNT([OrderID])*100
```

* Vizuallıq: `Gauge` və ya `Clustered Column Chart`.

### Power BI:

* Measure:

```DAX
Fill Rate (%) = DIVIDE(
    COUNTROWS(FILTER(Sales, Sales[Fullfilled]=1)),
    COUNTROWS(Sales)
) * 100
```

* Vizuallıq: `Gauge`, `Card` və ya `Column Chart`.

* Zamanla dəyişiklikləri göstərmək üçün **Line Chart** çox uyğundur.
