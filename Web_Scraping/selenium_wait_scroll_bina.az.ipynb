```python
import time

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.action_chains import ActionChains

class Scrolls:

    def __init__(self, driver, action):
        self.driver = driver
        self.action = action

    def scroll_by(self, x, y):
        self.driver.execute_script(f"window.scrollTo({x}, {y})")

    def scroll_to_bottom(self):
        self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    def scroll_to_top(self):
        self.driver.execute_script("window.scrollTo(0, 0)")

    def scroll_to_element(self, element):
        self.action.scroll_to_element(element).perform()
        self.driver.execute_script("""
        window.scrollTo({
            top: window.scrollY + 500,
        });
        """)
        


chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--start-maximized")
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
actions = ActionChains(driver)
scrolls = Scrolls(driver, actions)

wait = WebDriverWait(driver, 15, poll_frequency=1)

driver.get("https://bina.az/agsu/alqi-satqi/menziller")

FILTER_SELECT_LOCATOR = ("xpath", "(//span[@class='button'])[4]")
BAKI_SELECT_LOCATOR = ("xpath", "//span[text()='Bakı']")
ETRAFLI_SLECET_LOCATOR =("xpath", "//div[@class='search-row__cell search-row__cell--filters']")
METRO_SELECT_LOCATOR = ("xpath", "//span[text()='Metrostansiyalar']")
NARIMANOV_SELECET_LOCATOR = ("xpath", "//span[text()='Nəriman Nərimanov']")
AXTARISA_ELAVE_ET_LOCATOR = ("xpath", "(//div[text()='Axtarışa əlavə et'])[3]")
ELANI_GOSTER_LOCATOR = ("xpath", "//button[@class='search-filters__submit js-search-filters-items-count']")


FILTER_SELECT_BUTTON = driver.find_element(*FILTER_SELECT_LOCATOR)
BAKI_SELECT_BUTTON = driver.find_element(*BAKI_SELECT_LOCATOR)
ETRAFLI_BUTTON = driver.find_element(*ETRAFLI_SLECET_LOCATOR)
METRO_SELECT_BUTTON = driver.find_element(*METRO_SELECT_LOCATOR)
NARIMANOV_SELECT_BUTTON = driver.find_element(*NARIMANOV_SELECET_LOCATOR)
AXTARISA_ELAVE_ET_BUTTON = driver.find_element(*AXTARISA_ELAVE_ET_LOCATOR)
ELANI_GOSTER_BUTTON = driver.find_element(*ELANI_GOSTER_LOCATOR)


wait.until(EC.element_to_be_clickable(FILTER_SELECT_BUTTON)).click()
time.sleep(1)
wait.until(EC.element_to_be_clickable(BAKI_SELECT_BUTTON)).click()
time.sleep(1)
wait.until(EC.element_to_be_clickable(ETRAFLI_BUTTON)).click()
time.sleep(1)
wait.until(EC.element_to_be_clickable(METRO_SELECT_BUTTON)).click()
time.sleep(1)
wait.until(EC.element_to_be_clickable(NARIMANOV_SELECT_BUTTON)).click()
time.sleep(1)
wait.until(EC.element_to_be_clickable(AXTARISA_ELAVE_ET_BUTTON)).click()
time.sleep(1)
wait.until(EC.element_to_be_clickable(ELANI_GOSTER_BUTTON)).click()
# time.sleep(3)

price = []
location = []
otaq_sayi = []
sahe = []
mertebe = []

n = 0
while n<4:
    time.sleep(2)
    PRICE_LOCATOR = ("xpath", "//span[@class='price-val']")
    LOCATION_LOCATOR = ("xpath", "//div[@class='location']")
    OTAQ_LOCATOR = ("xpath", "//ul[@class='name']/li[1]")
    SAHE_LOCATOR = ("xpath", "//ul[@class='name']/li[2]")
    MERTEBE_LOCATOR = ("xpath", "//ul[@class='name']/li[3]")

    PRICE_BUTTON = driver.find_elements(*PRICE_LOCATOR)
    LOCATIO_BUTTON = driver.find_elements(*LOCATION_LOCATOR)
    OTAQ_BUTTON = driver.find_elements(*OTAQ_LOCATOR)
    SAHE_BUTTON = driver.find_elements(*SAHE_LOCATOR)
    MERTEBE_BUTTON = driver.find_elements(*MERTEBE_LOCATOR)

    
    print(len(PRICE_BUTTON))
    for i, j, z, t, m in zip(PRICE_BUTTON, LOCATIO_BUTTON, OTAQ_BUTTON, SAHE_BUTTON, MERTEBE_BUTTON):
        price.append(i.text)
        location.append(j.text)
        otaq_sayi.append(z.text)
        sahe.append(t.text)
        mertebe.append(m.text)
        
#     time.sleep(2)  
    
    NOVBETI_BUTTON = driver.find_element("xpath", "//span[@class='next']")
    scrolls.scroll_to_element(NOVBETI_BUTTON)
    
    try:
        NOVBETI_BUTTON = driver.find_element("xpath", "//span[@class='next']")
        wait.until(EC.element_to_be_clickable(NOVBETI_BUTTON)).click()
        

    except StaleElementReferenceException:
        print("Element köhnəlmişdir, yenidən tapmağa çalışıram...")
        time.sleep(2)  # Bir az gözləmək
    n+=1   

driver.quit()
import pandas as pd
pd.DataFrame({'Location': location, "Room": otaq_sayi, "Area": sahe, "Floor": mertebe, "Price": price})
```
